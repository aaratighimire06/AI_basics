```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('aarati.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize = (20,10))
plt.imshow(image)

```

```python
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
print(gray)
print(gray.shape)
```

```python
plt.figure(figsize = (20,10))
plt.imshow(gray,cmap = 'gray')
```

```python

# load cascade classifier 
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
faces = face_cascade.detectMultiScale(gray,1.3,5) # black and white image , ratio value for face , window size
print(faces)
print("no of faces",len(faces))
```

```python
for (x,y,w,h) in faces:
  image = cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),3) # image , height weidht haru , boundngbox, colour , window size  
plt.figure(figsize = (20,10))
plt.imshow(image)
```
