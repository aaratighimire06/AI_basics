```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#read image in BGR 
image = cv2.imread('aarati.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
filters = np.array([[-1,1,1],
                    [1,1,1],
                     [-1,1,-1]])
filtered_image = cv2.filter2D(image,-1,filters)

plt.imshow(filtered_image,cmap = 'gray')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#read image in BGR 
image = cv2.imread('aarati.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
filters =1/9*np.array([[1,1,1],
                    [1,1,1],
                     [1,1,1]])
filtered_image = cv2.filter2D(image,-1,filters)

plt.imshow(filtered_image,cmap = 'gray')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#read image in BGR 
image = cv2.imread('aarati.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
canney_image = cv2.Canny(image,80,240)
plt.imshow(canney_image,cmap = 'gray')
```
