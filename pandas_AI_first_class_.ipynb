```python
import pandas as pd
df = pd.read_csv('country_wise_latest.csv')
data = df[(df['Deaths'] > 20) & (df['Active']>1000)]
data.to_csv('new_covid_data.csv')
data
```

```python
country = list(data['Country/Region'])
active_cases = list(data['Active'])
sn = [i for i in range(1,len(country)+1)]
```

```python
import pandas as pd
data = {'Country':country,'Active Cases':active_cases,'Sn':sn}
df = pd.DataFrame(data)
```
